```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pytesseract
```

```python
doc = cv2.cvtColor(cv2.imread("../sample_forms/scanned.png"), cv2.COLOR_BGR2GRAY)

plt.imshow(doc, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
```

### Box out title

```python
width = doc.shape[1]
height = doc.shape[0]

# corners: top-left, bot-left, bot-right, top-right
xs = (width*0.2, width*0.2, width*0.8, width*0.8)
ys = (height*0.11, height*0.145, height*0.145, height*0.11)

plt.imshow(doc, cmap='gray', vmin=0, vmax=255)
plt.plot(xs+(xs[0],), ys+(ys[0],), linewidth=1)
plt.plot()
plt.axis('off')
```

```python
title = doc[int(height*0.11) : int(height*0.145), int(width*0.2):int(width*0.8)]

plt.imshow(title, cmap='gray', vmin=0, vmax=255)
plt.plot()
plt.axis('off')
```

### Do OCR on title

```python
# https://pypi.org/project/pytesseract/
title_text = pytesseract.image_to_string(title, lang='eng')
title_text
```
